```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from functions import model as m
import scipy.stats as stats
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
white = pd.read_csv('wines/WineQuality-WhiteWine.csv')
white.head()
```

```python
red = pd.read_csv('wines/WineQuality-RedWine.csv')
red.head()
```

```python
white['white'] = 1
white['red'] = 0
red['white'] = 0
red['red'] = 1
```

```python
df = pd.concat([white, red], ignore_index = True)
df.head()
```

```python
df.drop(columns =['red'], inplace=True)
df
```

```python
y = df[['quality']]
x = df.drop(['quality'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=16)
```

```python
import statsmodels.api as sm
```

```python
poisson = sm.GLM(y_train, x_train, family = sm.families.Poisson()).fit()
```

```python
poisson.summary()
```

```python
y_hat_train = poisson.predict(x_train)
y_hat_test = poisson.predict(x_test)
train_mse = mean_squared_error(y_train, y_hat_train)
test_mse = mean_squared_error(y_test, y_hat_test)
print('Train Root Mean Square Error:', train_mse**0.5)
print('Test Root Mean Square Error:', test_mse**0.5)
```

```python
#bootstrapping prediction interval
preds = []
n = x_train.shape[0]
nbootstrap = int(np.sqrt(n))

for i in range(nbootstrap):
    train_idx = np.random.choice(range(n), size = n, replace = True)
#     val_idx = np.array([idx for idx in range(n) if idx not in train_idx])
    model = sm.GLM(y_train[train_idx, :], x_train[train_idx, :], family = sm.families.Poisson()).fit()
    pred = model.predict(x_test[train_idx, :])
    preds.append(pred)
    
print(f'mean prediction: {np.mean(preds)}')
print(f'')
```
